```python
import os
import pickle
import pandas as pd
```

```python
data_dir = 'filtered_db_tab/'
files_to_read = os.listdir(data_dir)
print(files_to_read)
files_to_read.pop(4)
```

```python
for tab_data in files_to_read:
    print(tab_data)
    tab_dir = data_dir + tab_data
    df = pd.read_csv(tab_dir, delimiter='\t', header=None)
    print(df.head(3))
    c28_values = df[27].tolist()
    print('instances: ', len(c28_values))
    
    output_file_name = tab_dir.replace('.tabular', '.p')  # Change file extension to .p
    with open(output_file_name, 'wb') as file:
        pickle.dump(c28_values, file)
    print(f'{output_file_name} saved.')
    
```
